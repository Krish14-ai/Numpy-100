## 1. Arrays, Shapes & Dtypes

[NumPy absolute basics](https://numpy.org/doc/stable/user/absolute_beginners.html)

NumPy arrays are the core data structure for numerical computing in Python. Unlike Python lists, they are fixed-type and support fast vectorized operations.

**Worked example:** create an array and inspect its shape/dtype.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
print('shape:', arr.shape)
print('dtype:', arr.dtype)
```

## 3. Broadcasting & Vectorized Operations

[Broadcasting docs](https://numpy.org/doc/stable/user/basics.broadcasting.html)

Broadcasting lets NumPy apply operations between arrays of different (but compatible) shapes without writing explicit loops.

**Worked example:** subtract the column mean from each column of a matrix.

```python
m = np.random.rand(4, 3)
col_means = m.mean(axis=0)
centered = m - col_means
print(centered.mean(axis=0).round(6))
```

## 4. Aggregation, Sorting & Statistics

[Statistics docs](https://numpy.org/doc/stable/reference/routines.statistics.html)

NumPy provides fast reductions (sum, mean, std, min, max) along any axis, plus sorting and searching utilities.

**Worked example:** compute per-column statistics.

```python
data = np.random.randn(100, 3)
print('mean:', data.mean(axis=0))
print('std:', data.std(axis=0))
print('max:', data.max(axis=0))
```

## 5. Linear Algebra

[Linear algebra docs](https://numpy.org/doc/stable/reference/routines.linalg.html)

`numpy.linalg` provides matrix decompositions, determinants, inverses, eigenvalues, and more.

**Worked example:** solve a linear system Ax = b.

```python
A = np.array([[3, 1], [1, 2]], dtype=float)
b = np.array([9, 8], dtype=float)
x = np.linalg.solve(A, b)
print(x)
```

## 6. Random Sampling, Dates & Structured Arrays

[Random sampling docs](https://numpy.org/doc/stable/reference/random/index.html)

NumPy's random module (via `np.random.default_rng()`) supports many distributions, and NumPy also has native datetime and structured-array support.

**Worked example:** sample from a normal distribution with a reproducible generator.

```python
rng = np.random.default_rng(42)
sample = rng.normal(loc=0, scale=1, size=10)
print(sample.round(3))
```
